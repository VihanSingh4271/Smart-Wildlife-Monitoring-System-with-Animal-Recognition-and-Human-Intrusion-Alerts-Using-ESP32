```python
import cv2
import torch
from ultralytics import YOLO
```

```python
# Load the trained YOLO model
model = YOLO('/Users/ishanrawat/VihanProject/AnimalProject/best.pt')
```

```python
# Define class names based on your dataset
class_names = model.names  # Ensure this matches your classes in the dataset
```

```python
# Path to the input video
video_path = '/Users/ishanrawat/VihanProject/AnimalProject/forthtry.mp4'
output_path = '/Users/ishanrawat/VihanProject/AnimalProject/outputs/output_video.mp4'
```

```python
# Open video file
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print(f"Error: Unable to open video file {video_path}")
    exit()

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

print(f"Video dimensions: {width}x{height}, FPS: {fps}")
```

```python
# Initialize VideoWriter
fourcc = cv2.VideoWriter_fourcc(*'MP4V')  # Codec for .avi files
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

if not out.isOpened():
    print("Error: VideoWriter failed to open. Check codec or output path.")
    exit()

# Process video
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("End of video or error reading frame.")
        break

    # YOLO inference
    results = model.predict(frame, imgsz=640, conf=0.5, device="cpu")
    annotated_frame = results[0].plot()

    # Write frame to output file
    out.write(annotated_frame)

    # Display the frame
    cv2.imshow('YOLO Detection', annotated_frame)
    
    # Stop streaming when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Stopping video processing...")
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Output video saved to {output_path}")
```

```python
import cv2
from playsound import playsound  # Cross-platform sound playback library
from ultralytics import YOLO
import threading
```

```python
# Load the trained YOLO model
model = YOLO('/Users/ishanrawat/VihanProject/AnimalProject/best.pt')
```

```python
# Open the laptop camera
cap = cv2.VideoCapture(0)  # Use 0 for the default camera, or change to the appropriate camera index
if not cap.isOpened():
    print("Error: Unable to access the camera.")
    exit()

# Get camera properties
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
print(f"Camera resolution: {width}x{height}, FPS: {fps}")
```

```python
def play_alarm():
    try:
        print("Playing alarm sound...")
        playsound('/Users/ishanrawat/VihanProject/AnimalProject/alert.mp3')  # Replace with the path to your alarm sound file
    except Exception as e:
        print(f"Error playing sound: {e}")
```

```python
# Real-time detection
try:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Unable to read from camera.")
            break

        # YOLO inference
        results = model.predict(frame, imgsz=640, conf=0.5, device="cpu")
        annotated_frame = results[0].plot()

        # Print detected classes and labels for debugging
        print(f"Detected classes: {results[0].names}")

        # Check if any object detected is a human (based on class name or keyword)
        human_detected = False
        for result in results[0].boxes:  # Iterate through detected objects
            print(f"Detected label: {results[0].names[int(result.cls)]}")  # Print the detected class label for debugging
            # Adjust this condition to match your class names (e.g., "human" or something related)
            if "human" in results[0].names[int(result.cls)].lower():  # Check if the class label contains 'human'
                human_detected = True
                break

        if human_detected:
            print("Human detected! Triggering alarm...")
            alarm_thread = threading.Thread(target=play_alarm)
            alarm_thread.start()

        # Display the frame
        cv2.imshow('Real-Time Detection', annotated_frame)

        # Exit when 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("Exiting real-time detection...")
            break

except Exception as e:
    print(f"An error occurred: {e}")

# Release resources
finally:
    cap.release()
    cv2.destroyAllWindows()
```
